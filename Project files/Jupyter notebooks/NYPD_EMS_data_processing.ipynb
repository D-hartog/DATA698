```python
# Try and concatanate all files togeter
historic_data = pd.DataFrame()
for file in all_items: 
    df = pd.read_csv(folder_path + file, parse_dates = ["create_date", "incident_date", 
                                                   "add_ts","disp_ts", 
                                                   "arrivd_ts", "closng_ts"])
    historic_data = pd.concat([df,historic_data])
    print(f'Added {file} to data')
```
